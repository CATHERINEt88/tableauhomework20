```python
import pandas as pd
import datetime
import numpy as np
```

```python
#2019 data
june2019 = pd.read_csv("Data/JC-201906-citibike-tripdata.csv")
july2019 = pd.read_csv("Data/JC-201907-citibike-tripdata.csv")
aug2019 = pd.read_csv("Data/JC-201908-citibike-tripdata.csv")
sept2019 = pd.read_csv("Data/JC-201909-citibike-tripdata.csv")
oct2019 = pd.read_csv("Data/JC-201910-citibike-tripdata.csv")

#2020 data
june2020 = pd.read_csv("Data/JC-202006-citibike-tripdata.csv")
july2020 = pd.read_csv("Data/JC-202007-citibike-tripdata.csv")
aug2020 = pd.read_csv("Data/JC-202008-citibike-tripdata.csv")
sept2020 = pd.read_csv("Data/JC-202009-citibike-tripdata.csv")
oct2020 = pd.read_csv("Data/JC-202010-citibike-tripdata.csv")

#2021 data
june2021 = pd.read_csv("Data/JC-202106-citibike-tripdata.csv")
july2021 = pd.read_csv("Data/JC-202107-citibike-tripdata.csv")
aug2021 = pd.read_csv("Data/JC-202108-citibike-tripdata.csv")
sept2021 = pd.read_csv("Data/JC-202109-citibike-tripdata.csv")
oct2021 = pd.read_csv("Data/JC-202110-citibike-tripdata.csv")
```

```python
#combine data into one csv
frames = [june2019, july2019, aug2019, sept2019, oct2019, june2020, july2020, aug2020, sept2020, oct2020]
newpd = pd.concat(frames)
```

```python
newpd
```

```python
#drop the rows with birth year <1950 to exclude the possible outliers
indexname = newpd[newpd["birth year"]<1950].index
```

```python
new = newpd.drop(indexname,inplace=False)
```

```python
new
```

```python
#update the gender column, 1 as male, 2 as female, 0 as unknown
new["gender"].replace({1:"Male", 2:"Female", 0:"Unknown"}, inplace=True)
```

```python
new
```

```python
#extract the YEAR and the MONTH 
new["Trip year"] = pd.DatetimeIndex(new["starttime"]).year
```

```python
new["Trip month"] = pd.DatetimeIndex(new["starttime"]).month
```

```python
#save to new csv
new.to_csv("Data/combinedata.csv")
```
